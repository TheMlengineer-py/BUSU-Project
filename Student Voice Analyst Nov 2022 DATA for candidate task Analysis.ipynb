# Missing Value Analysis

```python
#missing value analysis
#The following isnull function will figure out if there are any missing values in the dataframe, and will then sum up the total for each column.
SpeakWeek.isnull().sum()
```

```python
#Missing value heatmap
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
sns.heatmap(SpeakWeek.isnull(),cbar=False,cmap='YlGnBu')
plt.ioff()
```

```python
#missing data plot
import warnings
import warnings
warnings.filterwarnings('ignore')
missing_SpeakWeek = SpeakWeek.isnull().sum(axis=0).reset_index()
missing_SpeakWeek.columns = ['column_name','missing_count']
missing_SpeakWeek = missing_SpeakWeek.loc[missing_SpeakWeek['missing_count']>0]
missing_SpeakWeek = missing_SpeakWeek.sort_values(by='missing_count')

ind = np.arange(missing_SpeakWeek.shape[0])
width = 0.5
fig,ax = plt.subplots(figsize=(4,10))
rects = ax.barh(ind,missing_SpeakWeek.missing_count.values,color='blue')
ax.set_yticks(ind)
ax.set_yticklabels(missing_SpeakWeek.column_name.values, rotation='horizontal')
ax.set_xlabel("Count of missing values")
ax.set_title("Number of missing values in each column")
plt.show()
```

```python
#Missing Data/Value percentage
#Using Missing Data percentage list
features_with_na=[features for features in df.columns if df[features].isnull().sum()]
for feature in features_with_na:
    print(feature, np.round(df[feature].isnull().mean(), 4), '% missing values.')
```

# Sentiment Analysis of Q5 (All Comments) using wordcloud and nltk modules 

```python
#Sentiment Analysis of Q5
#create some wordclouds to see the most frequently used words in the .
import re
import nltk
nltk.download('stopwords')
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
# Create stopword list:
stopwords = nltk.corpus.stopwords.words('english')
stopwords.append('newWord')
textt = " ".join(review for review in df.Q5)
#wordcloud = WordCloud(stopwords=stopwords).generate(textt)
wordcloud = WordCloud(stopwords=stopwords,background_color='white',mode="RGB", width=2000, height=1000).generate(textt)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('wordcloud11.png')
plt.show()
```
